Loading the cleaned dataset

```python
import pandas as pd

datasetFilePath = "/content/drive/MyDrive/Colab Notebooks/Gen AI/Mini Project 1/cleaned_dataset.csv"
structuredCleanedData = pd.read_csv(datasetFilePath)
```

Dividing the dataset into Training/Validation/Test (70/15/15) split

```python
from sklearn.model_selection import train_test_split


trainingData, tempData = train_test_split(structuredCleanedData, test_size=0.30, random_state=42, shuffle=True)
validationData, testingData = train_test_split(tempData, test_size=0.50, random_state=42, shuffle=True)

len(trainingData), len(validationData), len(testingData)
```

Converted the splitted corpus into Pandas dataset.

```python
from datasets import Dataset


summaryOutputColumn = "target"
summaryInputColumn = "source"

trainingDataset = Dataset.from_pandas(trainingData[[summaryInputColumn,summaryOutputColumn]].reset_index(drop=True))
validationDataset = Dataset.from_pandas(validationData[[summaryInputColumn,summaryOutputColumn]].reset_index(drop=True))
testingDataset = Dataset.from_pandas(testingData[[summaryInputColumn,summaryOutputColumn]].reset_index(drop=True))
```

Writing Training, Validation, & Test dataset into the drive

```python
import os


destinationFolderPath = "/content/drive/MyDrive/Colab Notebooks/Gen AI/Mini Project 1/"

trainingDataset.to_csv(destinationFolderPath+"training_dataset.csv", index=False)
print(os.path.exists(destinationFolderPath+"training_dataset.csv"))  # Should return True

validationDataset.to_csv(destinationFolderPath+"validation_dataset.csv", index=False)
print(os.path.exists(destinationFolderPath+"validation_dataset.csv"))  # Should return True

testingDataset.to_csv(destinationFolderPath+"test_dataset.csv", index=False)
print(os.path.exists(destinationFolderPath+"test_dataset.csv"))  # Should return True
```
